# SWFFT
the code necessary to run the Hardware Accelerated Cosmology Code (HACC) 3D distributed memory discrete fast Fourier transform, a light
weight make system, drivers for an example test code in C++ (TestDfft)as well as Fortran (TestFDfft), and a utility that checks grid si
zes and MPI rank layouts (CheckDecomposition).
